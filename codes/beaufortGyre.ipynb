### Spatial Structure and Timeseries Plots

### PreProcessing Cell

```python
# =============================================================================
# CELL 1 — PREPROCESSING
# Load all gridU and gridV files, extract relevant variables, select depth levels
# =============================================================================
 
import glob
import numpy as np
import xarray as xr
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from pathlib import Path
 
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.serif']  = ['cmr10', 'Computer Modern Roman', 'DejaVu Serif']
matplotlib.rcParams['axes.formatter.use_mathtext'] = True
 
# ── paths ─────────────────────────────────────────────────────────────────────
DATA_PATH  = Path("/mnt/storage7/clark/NEMO/Simulations/ARC60/Model_Runs/ARC60-ECP004")
OUTPUT_DIR = Path("../figures")
 
# ── year range to load (keep flexible) ───────────────────────────────────────
YEAR_START = 1993
YEAR_END   = 1999   # change to 2000+ when needed
 
# ── target depth levels (m) ───────────────────────────────────────────────────
TARGET_DEPTHS = [0, 50, 100, 200]   # surface + nominal depths
 
# ── file discovery ────────────────────────────────────────────────────────────
def _year_filter(files, y0, y1):
    """Keep only files whose filename year falls in [y0, y1]."""
    out = []
    for f in files:
        name = Path(f).stem          # e.g. ARC60-ECP004_y1995m03d08_gridU
        try:
            year = int(name.split('_y')[1][:4])
            if y0 <= year <= y1:
                out.append(f)
        except (IndexError, ValueError):
            pass
    return sorted(out)
 
all_u = glob.glob(str(DATA_PATH / "gridU" / "*_gridU.nc"))
all_v = glob.glob(str(DATA_PATH / "gridV" / "*_gridV.nc"))
 
u_files = _year_filter(all_u, YEAR_START, YEAR_END)
v_files = _year_filter(all_v, YEAR_START, YEAR_END)
 
print(f"Found {len(u_files)} gridU files  |  {len(v_files)} gridV files")
 
# ── preprocessing hooks (keep only what we need) ─────────────────────────────
def _preprocess_u(ds):
    return ds[['vozocrtx', 'nav_lat', 'nav_lon']]
 
def _preprocess_v(ds):
    return ds[['vomecrty', 'nav_lat', 'nav_lon']]
 
# ── lazy load ─────────────────────────────────────────────────────────────────
ds_u = xr.open_mfdataset(
    u_files,
    preprocess=_preprocess_u,
    combine='nested',
    concat_dim='time_counter',
    chunks={'time_counter': 10, 'depthu': 50, 'y': 500, 'x': 500},
    parallel=True,
)
 
ds_v = xr.open_mfdataset(
    v_files,
    preprocess=_preprocess_v,
    combine='nested',
    concat_dim='time_counter',
    chunks={'time_counter': 10, 'depthv': 50, 'y': 500, 'x': 500},
    parallel=True,
)
 
# ── select nearest depth levels ───────────────────────────────────────────────
#    .sel with method='nearest' picks the actual model level closest to each target
u_depths = ds_u['vozocrtx'].sel(depthu=TARGET_DEPTHS, method='nearest')
v_depths = ds_v['vomecrty'].sel(depthv=TARGET_DEPTHS, method='nearest')
 
# Rename depth dims to a common name so they align later
u_depths = u_depths.rename({'depthu': 'depth'})
v_depths = v_depths.rename({'depthv': 'depth'})
 
# Carry nav_lat / nav_lon from U grid (will be same for T-grid approx.)
nav_lat = ds_u['nav_lat']
nav_lon = ds_u['nav_lon']
 
print("u_depths shape:", u_depths.sizes)
print("v_depths shape:", v_depths.sizes)
print("Depth levels selected (U):", u_depths['depth'].values, "m")
print("Depth levels selected (V):", v_depths['depth'].values, "m")
```

### SpeedField Function

```python
# =============================================================================
# CELL 2 — SPEED FIELD
# Interpolate U/V to T-grid cell centres, compute speed, store lazily
# =============================================================================
 
def SpeedField(u_depths, v_depths):
    """
    Interpolate U (east/west faces) and V (north/south faces) to T-grid cell
    centres, then compute speed = sqrt(U² + V²).
 
    On a NEMO C-grid:
      T[i,j] = 0.5 * (U[i,j] + U[i-1,j])   → roll U by +1 along x
      T[i,j] = 0.5 * (V[i,j] + V[i,j-1])   → roll V by +1 along y
 
    Everything remains lazy (dask-backed) until .compute() is called.
 
    Parameters
    ----------
    u_depths : xr.DataArray  (time_counter, depth, y, x)
    v_depths : xr.DataArray  (time_counter, depth, y, x)
 
    Returns
    -------
    speeds : xr.DataArray  (time_counter, depth, y, x)  — speed in m/s
    """
    # Interpolate U → T  (average with the point one step west, i.e. roll x by 1)
    u_t = 0.5 * (u_depths + u_depths.roll(x=1, roll_coords=False))
 
    # Interpolate V → T  (average with the point one step south, i.e. roll y by 1)
    v_t = 0.5 * (v_depths + v_depths.roll(y=1, roll_coords=False))
 
    # Speed magnitude
    speeds = np.sqrt(u_t ** 2 + v_t ** 2)
    speeds.name = 'speed'
    speeds.attrs = {
        'long_name' : 'Ocean current speed',
        'units'     : 'm/s',
        'note'      : 'U and V interpolated to T-grid cell centres',
    }
 
    return speeds   # still lazy — no .compute() here
 
# Build the lazy speed DataArray across all dates and depths
speeds = SpeedField(u_depths, v_depths)
print("Speed field built (lazy):", speeds.sizes)
```

### Time Averages and Plotting

```python
# =============================================================================
# CELL 3 — TIME AVERAGES  &  PLOTTING
# =============================================================================
 
# ── season definitions (month numbers) ───────────────────────────────────────
SEASONS = {
    'JFM': [1, 2, 3],
    'AMJ': [4, 5, 6],
    'JAS': [7, 8, 9],
    'OND': [10, 11, 12],
}
 
def timeAverages(speeds, timePeriod):
    """
    Compute a time-mean speed field.
 
    Parameters
    ----------
    speeds     : xr.DataArray (time_counter, depth, y, x)
    timePeriod : int or str
        • int  → calendar year, e.g. 1994
        • str  → season code, one of 'JFM', 'AMJ', 'JAS', 'OND'
        • 'YXXXMXXDXX' string (e.g. '1994m03d08') → single-date pass-through
 
    Returns
    -------
    xr.DataArray — mean speed for the requested period (depth, y, x)
    """
    if isinstance(timePeriod, int):
        # ── year average ──────────────────────────────────────────────────────
        mask = speeds['time_counter'].dt.year == timePeriod
        subset = speeds.sel(time_counter=mask)
        if subset.sizes['time_counter'] == 0:
            raise ValueError(f"No data found for year {timePeriod}")
        result = subset.mean('time_counter')
        result.attrs['period'] = str(timePeriod)
        return result
 
    elif isinstance(timePeriod, str) and timePeriod in SEASONS:
        # ── seasonal average ──────────────────────────────────────────────────
        months = SEASONS[timePeriod]
        mask = speeds['time_counter'].dt.month.isin(months)
        subset = speeds.sel(time_counter=mask)
        if subset.sizes['time_counter'] == 0:
            raise ValueError(f"No data found for season {timePeriod}")
        result = subset.mean('time_counter')
        result.attrs['period'] = timePeriod
        return result
 
    else:
        raise ValueError(
            f"timePeriod must be an integer year or one of {list(SEASONS.keys())}. "
            f"Got: {timePeriod!r}"
        )
 
 
def speedGraph(
    speeds,
    nav_lat,
    nav_lon,
    speed_key,          # int year, season str, or xr.DataArray (pre-computed)
    depth,              # one of TARGET_DEPTHS in metres
    contours=50,        # number of contour lines; 0 → filled colormap only
    cmap='viridis',
    vmin=None,
    vmax=None,
    show=True,
    save=False,
    save_path=None,
):
    """
    Plot a speed map on a polar-stereographic projection.
 
    Parameters
    ----------
    speeds    : xr.DataArray (time_counter, depth, y, x) — lazy speed field
    nav_lat   : xr.DataArray (y, x)
    nav_lon   : xr.DataArray (y, x)
    speed_key : int | str | xr.DataArray
        • int  → year average (e.g. 1994)
        • str  → season average (e.g. 'JFM') or season+year (e.g. 'JFM_1994')
                 For a single date string use 'yYYYYmMMdDD' format.
        • xr.DataArray → use directly (must already be (depth, y, x))
    depth     : int — target depth in metres (must be in TARGET_DEPTHS)
    contours  : int — number of contour lines; 0 → pcolormesh only
    cmap      : str — matplotlib colormap name
    vmin/vmax : float or None — colour scale limits (auto if None)
    show      : bool
    save      : bool
    save_path : str or Path or None — file path for saving; auto-generated if None
    """
    # ── resolve the speed field ───────────────────────────────────────────────
    if isinstance(speed_key, xr.DataArray):
        field_2d = speed_key   # already (depth, y, x) or (y, x)
        label = speed_key.attrs.get('period', 'custom')
    else:
        field_2d = timeAverages(speeds, speed_key)
        label = str(speed_key)
 
    # ── select the requested depth ────────────────────────────────────────────
    if 'depth' in field_2d.dims:
        field_2d = field_2d.sel(depth=depth, method='nearest')
        actual_depth = float(field_2d['depth'].values)
    else:
        actual_depth = depth   # already 2-D
 
    # ── trigger computation (only the slice we need) ─────────────────────────
    print(f"Computing speed field for {label} at ~{actual_depth} m …")
    data = field_2d.compute().values
    lat  = nav_lat.values
    lon  = nav_lon.values
 
    # ── figure ───────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=(10, 10))
    
    pcm = ax.pcolormesh(
        lon, lat, data,
        cmap=cmap,
        vmin=_vmin, vmax=_vmax,
        shading='auto',
        rasterized=True,
    )
    
    if contours > 0:
        levels = np.linspace(_vmin, _vmax, contours)
        ax.contour(
            lon, lat, data,
            levels=levels,
            colors='k',
            linewidths=0.3,
            alpha=0.4,
        )
    
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
 
    # ── colorbar ─────────────────────────────────────────────────────────────
    cbar = fig.colorbar(pcm, ax=ax, orientation='vertical',
                        fraction=0.03, pad=0.08, shrink=0.7)
    cbar.set_label('Speed (m s$^{-1}$)', fontsize=11)
 
    # ── title ─────────────────────────────────────────────────────────────────
    depth_label = 'Surface' if actual_depth < 5 else f'{actual_depth:.0f} m'
    plot_type   = 'Colormap' if contours == 0 else f'{contours}-contour'
    ax.set_title(
        f'ARC60  —  Ocean Current Speed\n'
        f'Period: {label}   |   Depth: {depth_label}   |   {plot_type}',
        fontsize=12, pad=12,
    )
 
    plt.tight_layout()
 
    # ── save ──────────────────────────────────────────────────────────────────
    if save:
        if save_path is None:
            depth_str = 'surf' if actual_depth < 5 else f'{actual_depth:.0f}m'
            fname = OUTPUT_DIR / f"speed_{label}_{depth_str}_c{contours}.png"
        else:
            fname = Path(save_path)
        fig.savefig(fname, dpi=150, bbox_inches='tight')
        print(f"Figure saved → {fname}")
 
    if show:
        plt.show()
    else:
        plt.close(fig)
 
    return fig, ax
```
