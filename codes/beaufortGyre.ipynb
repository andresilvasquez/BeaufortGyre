### Spatial Structure and Timeseries Plots

### PreProcessing Cell

```python
# =============================================================================
# CELL 1 — SETUP
# Imports, paths, file index, nav_lat/lon, depth level mapping
# =============================================================================
 
import glob
import re
import numpy as np
import xarray as xr
import matplotlib
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm import tqdm
from datetime import datetime
 
matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.serif']  = ['cmr10', 'Computer Modern Roman', 'DejaVu Serif']
matplotlib.rcParams['axes.formatter.use_mathtext'] = True
 
# ── paths ─────────────────────────────────────────────────────────────────────
DATA_PATH      = Path("/mnt/storage7/clark/NEMO/Simulations/ARC60/Model_Runs/ARC60-ECP004")
PROCESSED_DIR  = Path("../processedData")
FIGURES_DIR    = Path("../figures")
PROCESSED_DIR.mkdir(exist_ok=True)
FIGURES_DIR.mkdir(exist_ok=True)
 
# ── year range (flexible) ─────────────────────────────────────────────────────
YEAR_START = 1993
YEAR_END   = 1999    # extend to 2000+ when needed
 
# ── target depths and integration boundaries ──────────────────────────────────
TARGET_DEPTHS = [0, 50, 100, 200]
 
# integrated=True averages velocities over these depth ranges
# depth=0 is invalid for integrated mode
DEPTH_BOUNDARIES = {
    50:  (0,   50),
    100: (50,  100),
    200: (100, 200),
}
 
# ── season definitions ────────────────────────────────────────────────────────
SEASONS = {
    'JFM': [1, 2, 3],
    'AMJ': [4, 5, 6],
    'JAS': [7, 8, 9],
    'OND': [10, 11, 12],
}
 
# ── file discovery ────────────────────────────────────────────────────────────
_DATE_RE = re.compile(r'_y(\d{4})m(\d{2})d(\d{2})_grid[UV]')
 
def _parse_date(filepath):
    m = _DATE_RE.search(Path(filepath).name)
    if m:
        return int(m.group(1)), int(m.group(2)), int(m.group(3))
    return None
 
def _build_file_index(files, y0, y1):
    index = {}
    for f in files:
        parsed = _parse_date(f)
        if parsed:
            y, mo, d = parsed
            if y0 <= y <= y1:
                index[datetime(y, mo, d)] = f
    return dict(sorted(index.items()))
 
all_u = glob.glob(str(DATA_PATH / "gridU" / "*_gridU.nc"))
all_v = glob.glob(str(DATA_PATH / "gridV" / "*_gridV.nc"))
 
u_index = _build_file_index(all_u, YEAR_START, YEAR_END)
v_index = _build_file_index(all_v, YEAR_START, YEAR_END)
 
common_dates = sorted(set(u_index) & set(v_index))
print(f"Found {len(common_dates)} matched U/V date pairs")
print(f"  First: {common_dates[0].date()}   Last: {common_dates[-1].date()}")
 
# ── load nav_lat/lon and depth levels once from first file ────────────────────
with xr.open_dataset(list(u_index.values())[0]) as ds:
    nav_lat            = ds['nav_lat'].load()        # (y, x)
    nav_lon            = ds['nav_lon'].load()        # (y, x)
    actual_depth_u     = ds['depthu'].values         # (50,)
 
print(f"\nGrid shape   : {nav_lat.shape}")
print(f"Depth levels : {len(actual_depth_u)}")
 
# ── single-level depth indices (non-integrated mode) ─────────────────────────
print("\nSingle-level depth mapping:")
single_depth_idx = {}
for target in TARGET_DEPTHS:
    idx = int(np.argmin(np.abs(actual_depth_u - target)))
    single_depth_idx[target] = idx
    print(f"  Target {target:>4} m  →  level index {idx:2d}  "
          f"(actual: {actual_depth_u[idx]:.2f} m)")
 
# ── integrated depth indices (integrated mode) ────────────────────────────────
print("\nIntegrated depth range mapping:")
integrated_depth_idx = {}
for target, (lower, upper) in DEPTH_BOUNDARIES.items():
    # depth=50  → 0   <= z <= 50
    # depth=100 → 50  <  z <= 100
    # depth=200 → 100 <  z <= 200
    if lower == 0:
        mask = (actual_depth_u >= lower) & (actual_depth_u <= upper)
    else:
        mask = (actual_depth_u >  lower) & (actual_depth_u <= upper)
    idx = np.where(mask)[0]
    integrated_depth_idx[target] = idx
    print(f"  depth={target:>3}m ({lower}–{upper} m)  →  "
          f"{len(idx)} levels  "
          f"[{actual_depth_u[idx[0]]:.1f} – {actual_depth_u[idx[-1]]:.1f} m]")
 
```

### SpeedField Function

```python
# =============================================================================
# CELL 2 — SINGLE FILE PREPROCESSING + SPEED FIELD
# =============================================================================
 
def SingleFilePreprocessing(date, depth, integrated=False):
    """
    Load one U/V file pair and return 2D velocity arrays at the given depth.
 
    Parameters
    ----------
    date       : datetime
    depth      : int — one of [0, 50, 100, 200]
    integrated : bool
        If True, return thickness-weighted vertical average over the depth range.
        integrated=True with depth=0 is invalid.
 
    Returns
    -------
    u, v : numpy arrays (y, x) in m/s, NaN over land
    """
    if depth not in TARGET_DEPTHS:
        raise ValueError(f"depth must be one of {TARGET_DEPTHS}, got {depth}")
    if integrated and depth == 0:
        raise ValueError(
            "integrated=True is not valid for depth=0 (surface has no layer to integrate). "
            "Use integrated=False for the surface."
        )
 
    uf = u_index[date]
    vf = v_index[date]
 
    if not integrated:
        # ── single level ──────────────────────────────────────────────────────
        d_idx = single_depth_idx[depth]
 
        with xr.open_dataset(uf) as ds:
            u = ds['vozocrtx'].isel(time_counter=0, depthu=d_idx).values  # (y, x)
        with xr.open_dataset(vf) as ds:
            v = ds['vomecrty'].isel(time_counter=0, depthv=d_idx).values  # (y, x)
 
        # mask fill values
        u[np.abs(u) > 1e10] = np.nan
        v[np.abs(v) > 1e10] = np.nan
 
    else:
        # ── thickness-weighted vertical average ───────────────────────────────
        idx = integrated_depth_idx[depth]   # array of level indices
 
        with xr.open_dataset(uf) as ds:
            u_3d = ds['vozocrtx'].isel(time_counter=0, depthu=idx).values  # (k, y, x)
            e3u  = ds['e3u'].isel(time_counter=0,      depthu=idx).values  # (k, y, x)
 
        with xr.open_dataset(vf) as ds:
            v_3d = ds['vomecrty'].isel(time_counter=0, depthv=idx).values
            e3v  = ds['e3v'].isel(time_counter=0,      depthv=idx).values
 
        # mask fill values
        u_3d[np.abs(u_3d) > 1e10] = np.nan
        v_3d[np.abs(v_3d) > 1e10] = np.nan
        e3u[ np.abs(e3u)  > 1e10] = np.nan
        e3v[ np.abs(e3v)  > 1e10] = np.nan
 
        # thickness-weighted average: Σ(u·e3) / Σ(e3)
        u = np.nansum(u_3d * e3u, axis=0) / np.nansum(e3u, axis=0)  # (y, x)
        v = np.nansum(v_3d * e3v, axis=0) / np.nansum(e3v, axis=0)
 
        # where all levels were NaN (land), result is 0/0 = NaN — already handled
        # but nansum(all-nan) = 0, so fix those explicitly
        all_nan_u = np.all(~np.isfinite(u_3d), axis=0)
        all_nan_v = np.all(~np.isfinite(v_3d), axis=0)
        u[all_nan_u] = np.nan
        v[all_nan_v] = np.nan
 
    return u.astype(np.float32), v.astype(np.float32)
 
 
def _parse_timeperiod(timePeriod):
    """
    Parse timePeriod string/int into (label, selected_dates).
 
    Accepts:
      'yYYYYmMMdDD'   → single date
      int             → full year average
      'JFM'/'AMJ'/... → seasonal average
    """
    if isinstance(timePeriod, str) and timePeriod.startswith('y'):
        m = re.match(r'y(\d{4})m(\d{2})d(\d{2})', timePeriod)
        if not m:
            raise ValueError(f"Date string must be 'yYYYYmMMdDD', got {timePeriod!r}")
        date = datetime(int(m.group(1)), int(m.group(2)), int(m.group(3)))
        if date not in u_index:
            raise ValueError(f"No file found for date {date.date()}")
        return timePeriod, [date]
 
    elif isinstance(timePeriod, int):
        selected = [d for d in common_dates if d.year == timePeriod]
        if not selected:
            raise ValueError(f"No dates found for year {timePeriod}")
        return str(timePeriod), selected
 
    elif isinstance(timePeriod, str) and timePeriod in SEASONS:
        months   = SEASONS[timePeriod]
        selected = [d for d in common_dates if d.month in months]
        if not selected:
            raise ValueError(f"No dates found for season {timePeriod}")
        return timePeriod, selected
 
    else:
        raise ValueError(
            f"timePeriod must be 'yYYYYmMMdDD', int year, "
            f"or one of {list(SEASONS.keys())}. Got: {timePeriod!r}"
        )
 
 
def speedField(timePeriod, depth, integrated=False):
    """
    Compute the mean speed field for a time period and depth level,
    then save it as a NetCDF file in ../processedData.
 
    Parameters
    ----------
    timePeriod : str or int
        'yYYYYmMMdDD' → single date
        int           → full year average
        'JFM' etc.    → seasonal average across all years
    depth      : int — one of [0, 50, 100, 200]
    integrated : bool — thickness-weighted average over depth range
 
    Returns
    -------
    out_path : Path — location of the saved NetCDF file
    """
    if integrated and depth == 0:
        raise ValueError("integrated=True is not valid for depth=0")
 
    label, selected_dates = _parse_timeperiod(timePeriod)
 
    suffix   = '_int' if integrated else ''
    fname    = f"speed_{label}_{depth}{suffix}.nc"
    out_path = PROCESSED_DIR / fname
 
    if out_path.exists():
        print(f"Already exists — skipping: {out_path}")
        return out_path
 
    print(f"Computing speed field")
    print(f"  Period     : {label}  ({len(selected_dates)} day(s))")
    print(f"  Depth      : {depth} m  |  integrated: {integrated}")
 
    # ── running accumulation (one file at a time — memory efficient) ──────────
    u_sum   = None
    v_sum   = None
    n_valid_u = None   # count of finite values per pixel
    n_valid_v = None
 
    for date in tqdm(selected_dates, desc=f"Loading {label}"):
        u, v = SingleFilePreprocessing(date, depth, integrated=integrated)
 
        u_fin = np.isfinite(u)
        v_fin = np.isfinite(v)
 
        if u_sum is None:
            u_sum     = np.where(u_fin, u, 0.0).astype(np.float64)
            v_sum     = np.where(v_fin, v, 0.0).astype(np.float64)
            n_valid_u = u_fin.astype(np.int32)
            n_valid_v = v_fin.astype(np.int32)
        else:
            u_sum     += np.where(u_fin, u, 0.0)
            v_sum     += np.where(v_fin, v, 0.0)
            n_valid_u += u_fin.astype(np.int32)
            n_valid_v += v_fin.astype(np.int32)
 
    # mean — only where we had valid data
    u_mean = np.where(n_valid_u > 0, u_sum / n_valid_u, np.nan).astype(np.float32)
    v_mean = np.where(n_valid_v > 0, v_sum / n_valid_v, np.nan).astype(np.float32)
 
    # ── C-grid interpolation to T-grid cell centres ───────────────────────────
    # U is on east  face: T[i,j] = 0.5*(U[i,j] + U[i-1,j]) → roll x (+1)
    # V is on north face: T[i,j] = 0.5*(V[i,j] + V[i,j-1]) → roll y (+1)
    u_t = 0.5 * (u_mean + np.roll(u_mean, 1, axis=-1))   # axis=-1 = x
    v_t = 0.5 * (v_mean + np.roll(v_mean, 1, axis=-2))   # axis=-2 = y
 
    # ── speed magnitude ───────────────────────────────────────────────────────
    speed = np.sqrt(u_t**2 + v_t**2).astype(np.float32)
 
    # propagate land mask: where both u and v are NaN → NaN speed
    land = ~np.isfinite(u_mean) & ~np.isfinite(v_mean)
    speed[land] = np.nan
 
    # ── build depth info string ───────────────────────────────────────────────
    if depth == 0:
        depth_info = f'Surface ({actual_depth_u[single_depth_idx[0]]:.1f} m)'
    elif integrated:
        lo, hi = DEPTH_BOUNDARIES[depth]
        depth_info = (f'Integrated {lo}–{hi} m  '
                      f'[{actual_depth_u[integrated_depth_idx[depth][0]]:.1f}–'
                      f'{actual_depth_u[integrated_depth_idx[depth][-1]]:.1f} m]')
    else:
        depth_info = (f'{depth} m target  '
                      f'(actual: {actual_depth_u[single_depth_idx[depth]]:.1f} m)')
 
    # ── save NetCDF ───────────────────────────────────────────────────────────
    ds_out = xr.Dataset(
        {
            'speed': xr.DataArray(
                speed,
                dims=['y', 'x'],
                attrs={
                    'long_name'  : 'Ocean current speed',
                    'units'      : 'm/s',
                    'depth_info' : depth_info,
                    'period'     : label,
                }
            ),
            'nav_lat': xr.DataArray(
                nav_lat.values, dims=['y', 'x'],
                attrs={'long_name': 'Latitude', 'units': 'degrees_north'}
            ),
            'nav_lon': xr.DataArray(
                nav_lon.values, dims=['y', 'x'],
                attrs={'long_name': 'Longitude', 'units': 'degrees_east'}
            ),
        },
        attrs={
            'title'      : f'ARC60 Ocean Current Speed — {label}',
            'depth_m'    : str(depth),
            'integrated' : str(integrated),
            'period'     : label,
            'n_days'     : str(len(selected_dates)),
            'created'    : str(datetime.now()),
        }
    )
 
    ds_out.to_netcdf(out_path)
    print(f"Saved → {out_path}")
    return out_path
 
 
# ── USAGE EXAMPLES ────────────────────────────────────────────────────────────
# Single date, surface:
# path = speedField('y1995m03d08', depth=0)
 
# Year average, 50 m:
# path = speedField(1995, depth=50)
 
# Season average, integrated 0–50 m:
# path = speedField('JFM', depth=50, integrated=True)
 
# Loop all years and depths:
# for year in range(1993, 2000):
#     for depth in TARGET_DEPTHS:
#         speedField(year, depth=depth)
#         if depth > 0:
#             speedField(year, depth=depth, integrated=True)
```

```python
# =============================================================================
# CELL 3 — GRAPH FUNCTION (Grid index space + lat/lon reference contours)
# =============================================================================

def graphFunction(
    speed_file,
    cmap='viridis',
    vmin=None,
    vmax=None,
    show=True,
    save=False,
    lat_step=5,    # degrees between latitude reference lines
    lon_step=30,   # degrees between longitude reference lines
):
    """
    Load a saved speed NetCDF from ../processedData and plot it
    in native grid index space, with lat/lon overlaid as reference contours.

    Parameters
    ----------
    speed_file : str or Path
    cmap       : str — matplotlib colormap
    vmin/vmax  : float or None — colour scale limits (auto if None)
    show       : bool
    save       : bool — saves to ../figures/
    lat_step   : spacing (degrees) between latitude contour lines
    lon_step   : spacing (degrees) between longitude contour lines
    """
    # ── resolve file path ─────────────────────────────────────────────────────
    path = Path(speed_file)
    if not path.is_absolute():
        path = PROCESSED_DIR / path
    if not path.exists():
        raise FileNotFoundError(
            f"Speed file not found: {path}\n"
            f"Available files:\n" +
            '\n'.join(f"  {f.name}" for f in sorted(PROCESSED_DIR.glob("*.nc")))
        )

    # ── load ──────────────────────────────────────────────────────────────────
    print(f"Loading {path.name} ...")
    with xr.open_dataset(path) as ds:
        data       = ds['speed'].values             # (y, x)
        lat        = ds['nav_lat'].values
        lon        = ds['nav_lon'].values
        label      = ds.attrs.get('period',      path.stem)
        integrated = ds.attrs.get('integrated',  'False') == 'True'
        depth_info = ds['speed'].attrs.get('depth_info', '?')

        print("lat min (raw, including fill):", lat.min())
        print("lat max (raw, including fill):", lat.max())
        print("unique low values:", np.unique(lat[lat < 1])[:10])

    # ── colour limits ─────────────────────────────────────────────────────────
    valid = data[np.isfinite(data)]
    if valid.size == 0:
        raise ValueError("Speed field contains no valid (non-NaN) data.")
    _vmin = vmin if vmin is not None else float(np.percentile(valid, 2))
    _vmax = vmax if vmax is not None else float(np.percentile(valid, 98))

    # ── figure ────────────────────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=(10, 10))

    pcm = ax.pcolormesh(
        data,
        cmap=cmap,
        vmin=_vmin, vmax=_vmax,
        shading='auto',
        rasterized=True,
    )

    # ── lat/lon reference contours ──────────────────────────────────────────
    lat_levels = np.arange(50, 91, lat_step)
    lon_levels = np.arange(-180, 181, lon_step)

    cs_lat = ax.contour(lat, levels=lat_levels, colors='white',
                        linewidths=0.6, alpha=0.7)
    ax.clabel(cs_lat, fmt='%d°N', fontsize=7, colors='white')

    cs_lon = ax.contour(lon, levels=lon_levels, colors='yellow',
                        linewidths=0.5, alpha=0.5)
    ax.clabel(cs_lon, fmt='%d°', fontsize=7, colors='yellow')

    ax.set_xlabel('x (grid index)', fontsize=11)
    ax.set_ylabel('y (grid index)', fontsize=11)

    # ── colorbar ─────────────────────────────────────────────────────────────
    cbar = fig.colorbar(pcm, ax=ax, orientation='vertical',
                        fraction=0.03, pad=0.04, shrink=0.8)
    cbar.set_label(r'Speed (m s$^{-1}$)', fontsize=11)

    # ── title ─────────────────────────────────────────────────────────────────
    int_str = ' depth-integrated' if integrated else ''
    ax.set_title(
        f'ARC60  —  Ocean Current Speed{int_str}\n'
        f'Period: {label}   |   {depth_info}',
        fontsize=12, pad=12,
    )

    plt.tight_layout()

    # ── save ──────────────────────────────────────────────────────────────────
    if save:
        fname = FIGURES_DIR / f"{path.stem}.png"
        fig.savefig(fname, dpi=150, bbox_inches='tight')
        print(f"Figure saved → {fname}")

    if show:
        plt.show()
    else:
        plt.close(fig)

    return fig, ax


# ── USAGE EXAMPLES ────────────────────────────────────────────────────────────
graphFunction('speed_y1995m03d08_0.nc', show=True, save=False)
```

```python
# =============================================================================
# CELL 3 — GRAPH FUNCTION (Cartopy Arctic Polar Projection, tripcolor method)
# =============================================================================

import cartopy.crs as ccrs
import cartopy.feature as cfeature

def graphFunction(
    speed_file,
    cmap='viridis',
    vmin=None,
    vmax=None,
    show=True,
    save=False,
    central_longitude=0,
    step=3,   # downsampling factor before triangulating (raise if too slow)
):
    """
    Load a saved speed NetCDF from ../processedData and plot it
    on a North Polar Stereographic projection using Cartopy,
    via tripcolor (triangulated point cloud) to avoid grid-fold artifacts.

    Parameters
    ----------
    speed_file          : str or Path
    cmap                 : str — matplotlib colormap
    vmin/vmax           : float or None — colour scale limits (auto if None)
    show                 : bool
    save                  : bool — saves to ../figures/
    central_longitude   : float — central longitude for projection (default 0)
    step                 : int — downsampling factor before triangulation
    """
    # ── resolve file path ─────────────────────────────────────────────────────
    path = Path(speed_file)
    if not path.is_absolute():
        path = PROCESSED_DIR / path
    if not path.exists():
        raise FileNotFoundError(
            f"Speed file not found: {path}\n"
            f"Available files:\n" +
            '\n'.join(f"  {f.name}" for f in sorted(PROCESSED_DIR.glob("*.nc")))
        )

    # ── load ──────────────────────────────────────────────────────────────────
    print(f"Loading {path.name} ...")
    with xr.open_dataset(path) as ds:
        data       = ds['speed'].values             # (y, x)
        lat        = ds['nav_lat'].values
        lon        = ds['nav_lon'].values
        label      = ds.attrs.get('period',      path.stem)
        integrated = ds.attrs.get('integrated',  'False') == 'True'
        depth_info = ds['speed'].attrs.get('depth_info', '?')

    # ── mask fill values (lat <= 0 marks land/no-data cells) ──────────────────
    fill_mask = (lat <= 0) | ~np.isfinite(data)

    # ── downsample before triangulating (tripcolor on 50M pts is too slow) ────
    lat_s  = lat[::step, ::step]
    lon_s  = lon[::step, ::step]
    data_s = data[::step, ::step]
    mask_s = fill_mask[::step, ::step]

    # ── flatten and keep only valid points ─────────────────────────────────────
    lat_flat  = lat_s[~mask_s]
    lon_flat  = lon_s[~mask_s]
    data_flat = data_s[~mask_s]

    if data_flat.size == 0:
        raise ValueError("Speed field contains no valid (non-NaN) data.")

    # ── colour limits ─────────────────────────────────────────────────────────
    _vmin = vmin if vmin is not None else float(np.percentile(data_flat, 2))
    _vmax = vmax if vmax is not None else float(np.percentile(data_flat, 98))

    # ── projection ────────────────────────────────────────────────────────────
    proj     = ccrs.NorthPolarStereo(central_longitude=central_longitude)
    data_crs = ccrs.PlateCarree()

    # ── figure ────────────────────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=(10, 10),
                           subplot_kw={'projection': proj})

    pcm = ax.tripcolor(
        lon_flat, lat_flat, data_flat,
        cmap=cmap,
        vmin=_vmin, vmax=_vmax,
        transform=data_crs,
    )

    # ── cartopy features ──────────────────────────────────────────────────────
    ax.add_feature(cfeature.LAND,      facecolor='lightgray', zorder=2)
    ax.add_feature(cfeature.COASTLINE, linewidth=0.5,          zorder=3)
    ax.add_feature(cfeature.BORDERS,   linewidth=0.3,          zorder=3)
    ax.gridlines(linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
    ax.set_extent([-180, 180, 50, 90], crs=ccrs.PlateCarree())

    # ── colorbar ──────────────────────────────────────────────────────────────
    cbar = fig.colorbar(pcm, ax=ax, orientation='vertical',
                        fraction=0.03, pad=0.04, shrink=0.8)
    cbar.set_label(r'Speed (m s$^{-1}$)', fontsize=11)

    # ── title ─────────────────────────────────────────────────────────────────
    int_str = ' — depth-integrated' if integrated else ''
    ax.set_title(
        f'ARC60  —  Ocean Current Speed{int_str}\n'
        f'Period: {label}   |   {depth_info}',
        fontsize=12, pad=12,
    )

    plt.tight_layout()

    # ── save ──────────────────────────────────────────────────────────────────
    if save:
        fname = FIGURES_DIR / f"{path.stem}.png"
        fig.savefig(fname, dpi=150, bbox_inches='tight')
        print(f"Figure saved → {fname}")

    if show:
        plt.show()
    else:
        plt.close(fig)

    return fig, ax


# ── USAGE EXAMPLES ────────────────────────────────────────────────────────────
graphFunction('speed_y1995m03d08_0.nc', show=True, save=False)
```
